Основные принципы, используемые для выявления аномалий:

1. Небольшое число первых главных компонент изменяется для нормальных данных.

2. Наименьшие компоненты для нормальных данных не изменяются.

3. Аномалии имеют в наименьших компонентах вариабельность.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Сгенерируем случайные данные

## инициируем новый генератор

```python
np.random.seed(42)
```

# Сгенерируем две выборки нормальных значений и объединим их как два кластера

```python
X_inliers = 0.3 * np.random.randn(100,2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]
```

```python
plt.hist(X_inliers)
plt.show()
```

# Cгенерируем некоторые аномалии (выбросы - outliers)

## Объединим два кластера (inliers и outliers)

```python
X_outliers = np.random.uniform(low = -4, high = 4, size = (20,2))
X = np.r_[X_inliers, X_outliers]
```

```python
plt.hist(X)
plt.show()
```
